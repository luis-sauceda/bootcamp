```python
import pandas as pd
```

```python
file = "Resources_movie_scores.csv"
```

```python
file_df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
file_df.head()
```

```python
df = file_df.set_index("FILM")
imdb_df = df.loc[:, ["IMDB", "IMDB_norm", "IMDB_norm_round", "IMDB_user_vote_count", ]]
imdb_df.head()
```

```python
only_good_df = imdb_df.loc[imdb_df["IMDB"] >= 7]
only_good_df.head()
```

```python
bad_movies_df = imdb_df.loc[imdb_df["IMDB_user_vote_count"] <= 1000]
bad_movies_df.head()
```

```python
imdb_df.to_csv("Salida_movies.csv", index = True, header = True)
```
