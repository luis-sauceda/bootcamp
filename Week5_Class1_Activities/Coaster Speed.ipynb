```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(0, 130, 10)
x_axis
```

```python
DangerDrop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
RailGun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
Danger_Handle = plt.plot(x_axis, DangerDrop, marker = "+", color ="green", label = "DangerDrop")
Rail_Handle = plt.plot(x_axis, RailGun, marker = "s", color ="red", label = "RailGun")
plt.legend(loc="upper right")
plt.set_title = "Roaler Coaster"
plt.xlabel("Time (seconds)")
plt.ylabel("Speed (MPH)")
plt.grid(True)
```
