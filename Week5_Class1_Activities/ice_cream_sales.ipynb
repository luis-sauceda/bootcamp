```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
temp = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
scoop_price =[values/random.random() for values in temp]
scoop_price
```

```python
# Tell matplotlib to create a scatter plot based upon the above data
plt.scatter(temp, sales, marker="o", facecolors="red", edgecolors="black", s=scoop_price, alpha=1)
```

```python
# Set the upper and lower limits of our y axis
plt.ylim(min(sales)-10, max(sales)+10)
```

```python
# Set the upper and lower limits of our x axis
plt.xlim(min(temp)-1,max(temp)+1)
```

```python
# Create a title, x label, and y label for our chart
plt.title("Relation between temperature and ice-cream sales")
plt.xlabel("Temperature (cÂ°)")
plt.ylabel("Sales (K USD)")
```
