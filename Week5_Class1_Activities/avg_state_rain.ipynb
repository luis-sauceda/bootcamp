```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
file = "Resources_avg_rain_state.csv"
rain_df=pd.read_csv(file, delimiter=",")
rain_df.columns
```

```python
# Set x axis and tick locations
states = rain_df["State"].values
tick_locations = [value for value in states]
states
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
millimetres = rain_df["Milli­metres"]
x_axis = np.arange(len(millimetres))
plt.bar(states, millimetres, color="r", alpha=0.5, align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, states)
```
