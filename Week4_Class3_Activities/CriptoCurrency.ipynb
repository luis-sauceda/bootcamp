```python
import pandas as pd
```

```python
bitcoin_csv = "Resources_bitcoin_cash_price.csv"
dash_csv = "Resources_dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
 bitcoin_df.head()
```

```python
 dash_df.head()
```

```python
merged_df = pd.merge(bitcoin_df, dash_df, on="Date")
merged_df.head()
```

```python
merged_df = merged_df.rename(columns = {"Open_x": "Open_bitcoin", "High_x": "High_bitcoin", "Low_x": "Low_bitcoin", 
                 "Close_x": "Close_bitcoin", "Volume_x": "Volume_bitcoin", "Market Cap_x": "Market_Cap_bitcoin",
                 "Open_y": "Open_dash", "High_y": "High_dash", "Low_y": "Low_dash", 
                 "Close_y": "Close_dash", "Volume_y": "Volume_dash", "Market Cap_y": "Market_Cap_dash"})
merged_df.head()
```

```python
bitcoin_open = merged_df["Open_bitcoin"].max()
dash_open = merged_df["Open_dash"].max()

# Collecting best close for Bitcoin and Dash
bitcoin_close = merged_df["Close_bitcoin"].max()
dash_close = merged_df["Close_dash"].max()

# Collecting the total volume for Bitcoin and Dash
bitcoin_volume = round(merged_df["Volume_bitcoin"].sum()/1000000, 2)
dash_volume = round(merged_df["Volume_dash"].sum()/1000000, 2)

summary_df = pd.DataFrame({"Best Bitcoin Open": [bitcoin_open],
                           "Best Bitcoin Close": [bitcoin_close],
                           "Total Bitcoin Volume": str(bitcoin_volume)+" million",
                           "Best Dash Open": [dash_open],
                           "Best Dash Close": [dash_close],
                           "Total Dash Volume": str(dash_volume)+" million"})

summary_df
```
