```python
# Dependencies
import matplotlib.pyplot as plt
from stats import median
import numpy as np
```

```python
### Data Points
arr = np.array([2.3, 10.2,11.2, 12.3, 14.5, 14.6, 15.0, 15.1, 19.0, 24.0])
arr
```

```python
# Find the median
median = median(arr)
median
```

```python
# Use numpy to create quartiles
q1 = np.percentile(arr,25)
q2 = np.percentile(arr,50)
q3 = np.percentile(arr,75)
```

```python
# Print the quartiles
print(q1)
print(q2)
print(q3)
```

```python
# Calculate the interquartile range
iqr = q3 - q1
print(iqr)
```

```python
# Find lower boundary
l_bound = q1 - 1.5 * iqr
print(l_bound)
```

```python
# Find upper boundary
u_bound =  q3 + 1.5 * iqr
print(u_bound)
```

```python
# Check for any lower outliers
arr[arr <= l_bound]
```

```python
# Check for any upper outliers
arr[arr >= u_bound]
```

```python
# Create box plot
plt.boxplot(arr, showmeans=True)
plt.grid()
plt.grid()
```
