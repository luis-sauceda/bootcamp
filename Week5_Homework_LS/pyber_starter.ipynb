```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Pyber_data_city_data.csv"
ride_data_to_load = "Pyber_data_ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset

merge_df = pd.merge(city_df, ride_df, on = "city", how = "outer")
# Display the data table for preview
merge_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#getting the max count of rides per city for x axis for each type of cities
urban_x = merge_df[merge_df["type"] == "Urban"].groupby("city").count().max()["ride_id"]
rural_x = merge_df[merge_df["type"] == "Rural"].groupby("city").count().max()["ride_id"]
suburban_x = merge_df[merge_df["type"] == "Suburban"].groupby("city").count().max()["ride_id"]
#print(urban_x, rural_x, suburban_x)

#print(urban_y, rural_y, suburban_y)
#getting the max average fare per city type for y axis
urban_y =  merge_df[merge_df["type"] == "Urban"].groupby("city").mean()["fare"].max()
rural_y =  merge_df[merge_df["type"] == "Rural"].groupby("city").mean()["fare"].max()
suburban_y =  merge_df[merge_df["type"] == "Suburban"].groupby("city").mean()["fare"].max()

# Build the scatter plots for each city types
#getting date average fare by city
urban_fares = merge_df[merge_df["type"] == "Urban"].groupby("city").mean()["fare"].values
rural_fares = merge_df[merge_df["type"] == "Rural"].groupby("city").mean()["fare"].values
suburban_fares =  merge_df[merge_df["type"] == "Suburban"].groupby("city").mean()["fare"].values

#getting total number of rides per city

urban_rides= merge_df[merge_df["type"] == "Urban"].groupby("city").count()["ride_id"].values
rural_rides = merge_df[merge_df["type"] == "Rural"].groupby("city").count()["ride_id"].values
suburban_rides = merge_df[merge_df["type"] == "Suburban"].groupby("city").count()["ride_id"].values

#getting the number of driver per city for the circle sizes
urban_drivers = merge_df[merge_df["type"] == "Urban"].groupby("city").count()["driver_count"].values
rural_drivers = merge_df[merge_df["type"] == "Rural"].groupby("city").count()["driver_count"].values
suburban_drivers = merge_df[merge_df["type"] == "Suburban"].groupby("city").count()["driver_count"].values

urban_hande = plt.scatter(urban_rides, urban_fares, s = urban_drivers*15, color ="coral", label = "Urban", 
                          alpha = 0.75, edgecolor = "black", linewidths = 1)
rural_hande = plt.scatter(rural_rides, rural_fares, s = rural_drivers*15, color ="gold", label = "Rural", 
                          alpha = 0.75, edgecolor = "black", linewidths = 1)
suburban_hande = plt.scatter(suburban_rides, suburban_fares, s = suburban_drivers*15, color ="skyblue", label = "Suburban", 
                             alpha = 0.75, edgecolor = "black", linewidths = 1)

# Incorporate the other graph properties
plt.legend(loc="upper right")
plt.set_title = "Roaler Coaster"
plt.xlabel("Total numbers of rides (per city)")
plt.ylabel("Average fare ($)")
plt.grid(True)

# Create a legend
plt.suptitle("Pyber Ride Sharing Data (2016)", fontsize =12)

# Incorporate a text label regarding circle size
plt.text(43, 35, 'Note:\n Circle size correlates with driver count per city ', style='italic',
        bbox={'facecolor': 'white', 'alpha': 0.5, 'pad': 10, 'edgecolor': 'none'})
# Save Figure
plt.savefig("PyberScatterFareVsRides.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_total_fare = merge_df[merge_df["type"] == "Urban"].groupby("type").sum()["fare"]
rural_total_fare = merge_df[merge_df["type"] == "Rural"].groupby("type").sum()["fare"]
suburban_total_fare = merge_df[merge_df["type"] == "Suburban"].groupby("type").sum()["fare"]

# Build Pie Chart
types = ["Urban", "Rural", "Suburban"]
types_percent = [int(urban_total_fare), int(rural_total_fare), int(suburban_total_fare)]
colors = ["coral", "gold", "skyblue"]
explode = (.1, 0, 0)

plt.pie(types_percent, explode=explode, labels= types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.suptitle("% of Total Fares by City Type", fontsize =12)
# Save Figure

plt.savefig("PyberPieFareByCityType.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_total_rides = merge_df[merge_df["type"] == "Urban"].groupby("type").count()["ride_id"]
rural_total_rides = merge_df[merge_df["type"] == "Rural"].groupby("type").count()["ride_id"]
suburban_total_rides = merge_df[merge_df["type"] == "Suburban"].groupby("type").count()["ride_id"]

# Build Pie Chart
types = ["Urban", "Rural", "Suburban"]
types_percent = [int(urban_total_rides), int(rural_total_rides), int(suburban_total_rides)]
colors = ["coral", "gold", "skyblue"]
explode = (.1, 0, 0)

plt.pie(types_percent, explode=explode, labels= types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.suptitle("% of Total Rides by City Type", fontsize =12)

# Save Figure
plt.savefig("PyberPieRidesByCityType.png")
plt.show()
```
